```python
from langchain_community.embeddings import FastEmbedEmbeddings

embeddings = FastEmbedEmbeddings(
    model_name="BAAI/bge-small-en-v1.5"
)
```

```python
embeddings.embed_query(pedacos[10].page_content)
```

```python
from langchain_community.vectorstores import InMemoryVectorStore

vectorstore = InMemoryVectorStore.from_documents(
    documents=pedacos[50:60],
    embedding=embeddings
)
```

```python
#Para usar esse banco de dados vetorial, precisamos de um buscador ou um retriever, que é o R do RAG:

retriever = vectorstore.as_retriever(search_kwargs={"k": 3}) #número de pedaços semelhantes a serem retornados
```

```python
query = "O que é processo penal?"
similar_chunks = retriever.invoke(query)
```

```python
#resgatar documentos similares

similar_texts = [doc.page_content for doc in similar_chunks]
similar_texts
```

```python
contexto = "\n\n".join(similar_texts)
```

```python
query = "O que é processo penal?"
query_embedding = embeddings.embed_query(query)
query_embedding
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "Responda usando exclusivamente os conteudo fornecido. \n\nContexto:\n{contexto}"),
    ("human", "{query}")
])
```

```python
from huggingface_hub import login
login()
```

```python
from huggingface_hub import whoami
whoami()
```

```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="facebook/opt-125m",
    max_new_tokens=128,
    do_sample=False
)
```

```python
from langchain_community.llms import HuggingFacePipeline

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
chain = prompt | llm
```

```python
resposta = chain.invoke({
    "contexto": contexto,
    "query": query
})
```
